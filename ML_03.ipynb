Q2: List down techniques used to handle missing data. Give an example of each with python code.
We generally handling missing values with mean median mode.We also drop the column or row but it cause data loss.


```python
import seaborn as sns
import pandas as pd 
import numpy as np
df=sns.load_dataset('titanic')
df.isnull().sum()
df.dropna()
df.dropna(axis=1)
df['new_age']=df['age'].fillna(df['age'].mean())
df['new_age']=df['age'].fillna(df['age'].median())
df['new_age']=df['age'].fillna(df['age'].mode()[0])

```

What are Up-sampling and Down-sampling? Explain with an example when up-sampling and down-
sampling are required.
Up sampling is the technique to introduce more datapoints in the minority class dataset for imbalanced dataset.
Down sampling is the technique to reduce  datapoints in the majority class dataset for imbalanced dataset.


```python
import seaborn as sns
import pandas as pd 
import numpy as np
from sklearn.utils import resample
n_sample=1000
n_class_0_w=0.9
n_class_0_l=int(n_class_0_w*n_sample)
n_class_1_l=n_sample-n_class_0_l
n_class_0=pd.DataFrame({'f1':np.random.normal(loc=0,scale=1,size=n_class_0_l),
                        'target':[0]*n_class_0_l})
n_class_1=pd.DataFrame({'f1':np.random.normal(loc=2,scale=1,size=n_class_1_l),
                        'target':[1]*n_class_1_l})
df=pd.concat([n_class_0,n_class_1])
df['target'].value_counts()

```

```python
df_min=df[df['target']==1]
df_maj=df[df['target']==0]
```

```python
df_maj=resample(df_min,n_samples=len(df_maj),replace=True)
```

```python
df_min=resample(df_maj,replace=False,n_samples=len(df_min))
```

Q10: When attempting to estimate customer satisfaction for a project, you discover that the dataset is
unbalanced, with the bulk of customers reporting being satisfied. What methods can you employ to
balance the dataset and down-sample the majority class?
Down sampling ,up sampling, smote

```python
import seaborn as sns
import pandas as pd 
import numpy as np
from imblearn.over_sampling import SMOTE
n_sample=1000
n_class_0_w=0.9
n_class_0_l=int(n_class_0_w*n_sample)
n_class_1_l=n_sample-n_class_0_l
n_class_0=pd.DataFrame({'f1':np.random.normal(loc=0,scale=1,size=n_class_0_l),
                        'target':[0]*n_class_0_l})
n_class_1=pd.DataFrame({'f1':np.random.normal(loc=2,scale=1,size=n_class_1_l),
                        'target':[1]*n_class_1_l})
df=pd.concat([n_class_0,n_class_1])
df['target'].value_counts()

smote=SMOTE()
X,y=smote.fit_resample(df[['f1']],df['target'])

```
