Glyphs in Bokeh are the visual shapes used to represent data points on a plot.

```python
from bokeh.plotting import figure, show

# Create a figure
p = figure(title="Bokeh Glyph Example")

# Add glyphs
p.circle([1, 2, 3, 4], [2, 4, 6, 8], size=10, color="blue", legend_label="Circles")
p.line([1, 2, 3, 4], [2, 4, 6, 8], line_width=2)

# Display plot
show(p)
```

```python
#Customize the Title
p.title.text = "Sales Growth"
p.title.align = "center"
p.title.text_font_size = "16pt"
#Customize the Axes
# X-axis
p.xaxis.axis_label = "Year"
p.xaxis.major_label_text_font_size = "12pt"
p.xaxis.axis_line_width = 2

# Y-axis
p.yaxis.axis_label = "Revenue"
p.yaxis.major_label_text_font_size = "12pt"
p.yaxis.axis_line_width = 2
#Customize the Legend
p.legend.title = "Legend"
p.legend.location = "top_left"
p.legend.label_text_font_size = "10pt"
```
