```python
a = 10
b=int(input("enter a number"))
try:
    s = a / b
except ZeroDivisionError as e:
    print("ohhh error")
else:
    print("ohh no error it seems")
finally:
    print("finally we have done it")
```

```python
a = 10
try:
    b = int(input("enter a number: "))
    if b == 0:
        raise ZeroDivisionError("Cannot divide by zero")
    s = a / b
except ZeroDivisionError as e:
    print("ohhh error")
else:
    print("ohh no error it seems")
finally:
    print("finally we have done it")
```

Custom exception are exception made by the user.They are important so that user can make there own exception.


```python
class valid(Exception):
    def validate_age(self,msg):
       self.msg=msg
    
```

```python
def val_age(age):
    if age<0:
        raise valid("oh wrong age guys")
    elif age>200:
        raise valid("oh again wrong age")
    else:
        print('age valid')
    
    
```

```python
try:
    age=int(input("enter the age"))
    val_age(age)
except valid as e:
    print(e)
```

```python
class bank(Exception):
    def account(self,balance,msg):
        self.msg=msg
        self.balance=balance
        
```

```python
def account_detail(balance , withdraw):
    if balance>withdraw:
        raise bank("withdraw")
        print("goood")
    else:
        raise bank("no withdraw")
    
```

```python
try:
        balance=int(input("enter the balance"))
        withdraw=int(input("enter the withdraw"))
        account_detail(balance,withdraw)
except bank as e:
    print(e)
```
