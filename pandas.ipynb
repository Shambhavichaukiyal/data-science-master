```python
import pandas as pd

# Create Series
name = pd.Series(['Alice', 'Bob', 'Claire'])
age = pd.Series([25, 30, 27])
gender = pd.Series(['Female', 'Male', 'Female'])

# Create DataFrame from multiple Series
df = pd.DataFrame({
    'Name': name,
    'Age': age,
    'Gender': gender
})

# Print DataFrame
print(df)
```
