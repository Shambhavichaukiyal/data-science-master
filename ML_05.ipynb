Q1. What is data encoding? How is it useful in data science?
Data encoding is the process of converting categorical datapoints to numeric datapoints.
It is used in data science because ML algorithms work in numeric data not in categorical data.

 What is nominal encoding? Provide an example of how you would use it in a real-world scenario.
 Nominal encoding is NOT an actual encoding algorithm.
It simply refers to handling nominal categorical data—that is, categories with no inherent order.
Color: Red, Blue, Green
City: Delhi, Mumbai, Pune

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
oe=OneHotEncoder()
df=pd.DataFrame({'color':['red','blue','green','violet','red','green']})
c=oe.fit_transform(df[['color']])
```

```python
pd.DataFrame(c.toarray(),columns=oe.get_feature_names_out())
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
oe=LabelEncoder()
df=pd.DataFrame({'color':['red','blue','green','violet','red','green']})
cs=oe.fit_transform(df[['color']])
cs

```
