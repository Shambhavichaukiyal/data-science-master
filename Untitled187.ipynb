Using the with statement with open() in Python ensures that files are automatically closed after use, even if exceptions occur, thereby preventing resource leaks and making the code cleaner and more robust.







The writelines() function is used to write a list of strings to a file
The write() function is used to write a string to a file

```python
f=open("textname1.txt","w")
f.write("gvftftytcftyf hgytfyt6 hfvytcyt")
f.close()
```

```python
f= open("textname1.txt", "r") 
print(f.read())
```

```python
f=open("textname1.txt","w")
lines = ["Line 1\n", "Line 2\n", "Line 3\n"]
f.writelines(lines)
f.close()
```

```python
f= open("textname1.txt", "r") 
print(f.read())
```
