Flask Framework is a library in python used for web application.Flask is simple wasy to use and the learning curve is easier compared to other framework.

```python
from flask import Flask

app=Flask(__name__)
@app.route("/")
def hello():
    return "hello world"
if __name__=="__main__":
    app.run(host="0.0.0.0")
```

App routing in flask is a way to map python code to url through flask.When a user visits a URL in the browser, Flask needs to know which function should handle that request.

```python
from flask import Flask

app=Flask(__name__)
@app.route("/welcome")
def hello():
    return "Welcome to ABC Corporation"
@app.route("/")
def val():
    data={"Company Name":"ABC Coporation","Location":"India","Phone Number":8979613025}
    return data
if __name__=="__main__":
    app.run(host="0.0.0.0")
```

url_for() is used in Flask for URL building.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route("/welcome")
def hello():
    return "Welcome to ABC Corporation"

@app.route("/")
def val():
    # Generate URLs dynamically using url_for
    welcome_url = url_for('hello')  # url_for uses the function name, not the route
    home_url = url_for('val')
    
    data = {
        "Company Name": "ABC Corporation",
        "Location": "India",
        "Phone Number": 8979613025,
        "Welcome Page URL": welcome_url,
        "Home Page URL": home_url
    }
    return data
if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000, debug=True, use_reloader=False)

```
