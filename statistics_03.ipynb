```python
import math

def normal_pdf(x, mean, std):
    pdf = (1 / (std * math.sqrt(2 * math.pi))) * \
          math.exp(-0.5 * ((x - mean) / std) ** 2)
    return pdf
result = normal_pdf(x=50, mean=50, std=10)
print(result)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10          # number of trials
p = 0.4         # probability of success
size = 1000     # sample size

# Generate random sample
data = np.random.binomial(n, p, size)

# Plot histogram
plt.figure()
plt.hist(data, bins=range(0, n + 2))
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Histogram of Binomial Distribution (n=10, p=0.4)")
plt.show()
```

```python
import math

def poisson_cdf(k, lam):
    cdf = 0
    for i in range(0, k + 1):
        cdf += (math.exp(-lam) * lam**i) / math.factorial(i)
    return cdf
result = poisson_cdf(k=3, lam=2.5)
print(result)
```

Number of trials

Binomial distribution:
Number of trials n is fixed (e.g., 10 coin tosses).

Poisson distribution:
Number of trials is not fixed; we count events in a time/space interval (e.g., calls per hour).

2. Type of outcome

Binomial:
Each trial has only two outcomes – success or failure.

Poisson:
Counts how many times an event occurs (0, 1, 2, 3, …).

3. Probability / parameter

Binomial:
Uses probability of success (p) for each trial.

Poisson:
Uses λ (lambda), the average number of events.

4. Parameters involved

Binomial:
Depends on n (number of trials) and p (probability).

Poisson:
Depends only on λ (mean rate).

5. Mean

Binomial:
Mean =np

Poisson:
Mean =λ

6. Variance

Binomial:
Variance =npq


Poisson:
Variance=λ

```python
import numpy as np

# Parameters
lam = 5          # mean of Poisson distribution
size = 1000      # sample size

# Generate random sample
data = np.random.poisson(lam, size)

# Calculate sample mean and variance
sample_mean = np.mean(data)
sample_variance = np.var(data)

print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
```
