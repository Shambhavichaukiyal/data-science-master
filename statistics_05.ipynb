Estimation statistics is the branch of statistics in which we generally estimate population parameters using sample data.
Point-In this estimation statistics we have one point of estimate population parameter using sample data.
interval-In this estimation statistics we have interval of estimate population parameter using sample data.

```python
import math

def estimate_population_mean(sample_mean, sample_std, sample_size):
    """
    Estimates the population mean using sample statistics.
    """
    population_mean_estimate = sample_mean
    standard_error = sample_std / math.sqrt(sample_size)
    
    return population_mean_estimate, standard_error
```

```python
mean, se = estimate_population_mean(50, 10, 100)
print("Estimated population mean:", mean)
print("Standard error:", se)
```

Hypothesis testing is a statistical method in which we make an assumption about a population parameter and then use sample data to test whether there is enough evidence to reject or contradict that assumption.

Check the validity of assumptions or claims about a population

Make decisions using data instead of guesswork

Compare sample results with expected or claimed values

Helps in decision making under uncertainty



null hypothesis-The average weight of male college students is less than or equal to the average weight of female college students.

Alternative Hypothesis (H₁):
The average weight of male college students is greater than the average weight of female college students.



```python
import numpy as np
from scipy import stats

# Sample data from two populations
sample1 = np.array([70, 72, 68, 75, 71])
sample2 = np.array([65, 67, 64, 66, 63])

# Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

# Significance level
alpha = 0.05

print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

def plot_t_distribution(df):
    """
    Plots a Student's t-distribution for the given degrees of freedom (df)
    """
    # Generate x values
    x = np.linspace(-5, 5, 500)
    
    # Compute t-distribution PDF
    y = t.pdf(x, df)
    
    # Plot
    plt.plot(x, y, label=f"t-distribution (df={df})", color='blue')
    plt.title("Student's t-Distribution")
    plt.xlabel("x")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example: plot with 11 degrees of freedom
plot_t_distribution(11)
```

```python
import numpy as np
from scipy import stats

# Step 1: Generate two random samples of equal size
np.random.seed(0)  # for reproducibility
sample_size = 30

# Sample 1: from population with mean=50, std=5
sample1 = np.random.normal(loc=50, scale=5, size=sample_size)

# Sample 2: from population with mean=52, std=5
sample2 = np.random.normal(loc=52, scale=5, size=sample_size)

# Step 2: Define null hypothesis
# H0: population means are equal (mu1 = mu2)
# H1: population means are not equal (mu1 != mu2)

# Step 3: Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

# Step 4: Set significance level
alpha = 0.05

# Step 5: Print results
print("Sample 1:", sample1)
print("Sample 2:", sample2)
print("\nT-statistic:", t_statistic)
print("P-value:", p_value)

# Step 6: Make decision
if p_value < alpha:
    print("Reject the null hypothesis → Means are significantly different")
else:
    print("Fail to reject the null hypothesis → No significant difference in means")
```
