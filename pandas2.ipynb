```python
import pandas as pd

# Take input from user
date1 = input("Enter first date (YYYY-MM-DD): ")
date2 = input("Enter second date (YYYY-MM-DD): ")

# Convert input strings to pandas Timestamp
d1 = pd.Timestamp(date1)
d2 = pd.Timestamp(date2)

# Calculate time difference
diff = d2 - d1

# Display the result
print("Difference in days:", diff.days)
print("Difference in hours:", diff.total_seconds() / 3600)
print("Difference in minutes:", diff.total_seconds() / 60)
```

```python
import pandas as pd

# Take inputs from user
file_path = input("Enter CSV file path: ")
column_name = input("Enter column name to convert to categorical: ")

# Take category order as comma-separated values
category_order = input(
    "Enter category order (comma-separated, e.g. low,medium,high): "
).split(",")

# Read CSV file
df = pd.read_csv(file_path)

# Convert column to categorical with specified order
df[column_name] = pd.Categorical(
    df[column_name],
    categories=category_order,
    ordered=True
)

# Sort data based on categorical column
df_sorted = df.sort_values(by=column_name)

# Display sorted data
print("\nSorted Data:")
print(df_sorted)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Take CSV file path from user
file_path = input("Enter the CSV file path: ")

# Read CSV file
df = pd.read_csv(file_path)

# Set Date column as index
df.set_index('Date', inplace=True)

# Plot stacked bar chart
df.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel("Time Period")
plt.ylabel("Sales")
plt.title("Sales of Product Categories Over Time")

# Display the chart
plt.show()
```

```python
import pandas as pd

# Take CSV file path from user
file_path = input("Enter the CSV file path: ")

# Read the CSV file
df = pd.read_csv(file_path)

# Calculate statistics
mean_score = df['Test_Score'].mean()
median_score = df['Test_Score'].median()
mode_score = df['Test_Score'].mode()[0]

# Create a result table
result = pd.DataFrame({
    'Statistic': ['Mean', 'Median', 'Mode'],
    'Test_Score': [mean_score, median_score, mode_score]
})

# Display the result table
print("\nStatistical Summary of Test Scores:")
print(result)
```
