```python
import numpy as np
list_ = ['1', '2', '3', '4', '5']
array_list = np.array(object = list_)
for i in list_:
    print(type(i))
```

```python
import numpy as np
list_ = ['1', '2', '3', '4', '5']
array_list = np.array(object = list_)
for i in array_list:
    print(type(i))
```

```python
import numpy as np
list_ = ['1', '2', '3', '4', '5']
array_list = np.array(object = list_, dtype = int)
for i in array_list:
    print(type(i))
```

First array_list was String but now i have mentioned to convert dtype to int so the datatype here is int.
