Multiprocessing can be defined as running multiple processing at the same time in  single thread.It is useful of performance efficiency.While multithreading can be defined as running multiple thread at the same time in single process.

Multiprocessing pool can be defined as the pool of processes that we have to run.It uses multiprocessing.Pool()  function inside function we defined the number of processes.It is used for overall efficiency improve performance and we do not need to create and manage processes.

```python
'''Multiprocessing pool can be defined as the pool of processes that we have to run.It uses multiprocessing.Pool()  function inside function we defined the number of processes.It is used for overall efficiency improve performance and we do not need to create and manage processes.'''
import multiprocessing
def sq(a):
    return a*a
if __name__=="__main__":
    with multiprocessing.Pool(processes=4) as pool:
        out=pool.map(sq,[1,2,3,4])
        print(out)
```
