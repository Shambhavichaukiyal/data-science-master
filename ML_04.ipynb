 What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an
example to illustrate its application.

PCA is a feature extraction technique that is used for reduce the feature from high dimenion to low dimension.PCA usually do dimensionality reduction by combining important correlate features for best acurracy and less loss of datapoints.

```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

df=sns.load_dataset('tips')
x = df[['total_bill', 'tip']]
x_new=sc.fit_transform(x)
from sklearn.decomposition import PCA
pca=PCA()
y=pca.fit_transform(x_new)
```

In a food delivery recommendation system, features like price, rating, and delivery time are on very different scales:
Price → ₹100 to ₹1000
Rating → 1 to 5
Delivery time → 10 to 60 minutes

If we use them directly, features with larger values (like price) will dominate the model.
Min max scaler is a feature scaling technique so that datapoints have fixed value in range 0 -1 .

How to do it -
Select numerical features: price, rating, delivery time

Compute the minimum and maximum for each feature

Apply Min-Max scaling column-wise

Use the scaled features for similarity calculation or model training
