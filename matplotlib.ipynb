Matplotlib is a library in python used for visualization of data.It is used for visualization of data.Line plot,scatter plot,bar plot,hist,pie.

```python
#Scatter plot a visulization in which all the data are marked in x,y by point.
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.plot(x,y)
plt.xlabel("The X-Axis")
plt.ylabel("The Y-Axis")
plt.title("The graph for use")
```

subplot function is used in greater dimension graph than 2d fo visualization.

```python
import numpy as np
import matplotlib.pyplot as plt

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.plot(x1, y1, label="Line 1")
plt.plot(x2, y2, label="Line 2")
plt.plot(x3, y3, label="Line 3")
plt.plot(x4, y4, label="Line 4")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()
plt.show()
```

A bar plot (bar chart) is a graphical representation of data where categorical values are shown on one axis and numerical values are shown using rectangular bars.Compare different categories.

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("The Company")
plt.ylabel("The profit")
plt.title("Company Profit")
```

```python
plt.barh(company, profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Company Profit (Horizontal Bar Plot)")
plt.show()
```

A box plot (also called a box-and-whisker plot) is a statistical graph that displays the distribution of data

```python
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2], labels=["Box 1", "Box 2"])
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```
