We intend to use Exception class while making a custom function so that it inherit the property of the exception class for efficiency consistency and what not.

```python
def print_exception_hierarchy(exception, indent=0):
    print("  " * indent + exception.__name__)
    for base_exception in exception.__bases__:
        print_exception_hierarchy(base_exception, indent + 1)

print("Python Exception Hierarchy:")
print_exception_hierarchy(BaseException)
```

Lookup Function is used for index related issue in the program .


```python

l=[1,2,3]
try:
    a=l[3]
except IndexError as e:
    print("error")
    
```

```python
l = {1: "fgf", 2: "Ddcv", 3: "dff"}
try:
    a = l[5] 
except KeyError as e:
    print("error")  
```
