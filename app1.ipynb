GET Method
Used to request or retrieve data from the server.

Post method
Used to send data to server.

Request is used to retrieve data from user.

Redirect is used to send user from one url to another.

Template are html code that are used to design interface for the application.
render_template() is used to call html template.

```python
from flask import Flask,request,render_template ,jsonify
app=Flask(__name__)
@app.route("/")
def homepage():
    return "Working fine"


    
@app.route("/mathpost",methods=['POST'])
def operate1():
    if request.method=="POST":
        ops=request.json['operation']
        num1=int(request.json['num1'])
        num2=int(request.json['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + 'and ' + str(num2) + " is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + 'and ' + str(num2) + " is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + 'and ' + str(num2) + " is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + 'and ' + str(num2) + " is " + str(r)
        return jsonify(result)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
