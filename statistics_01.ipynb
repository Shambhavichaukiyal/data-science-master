Understanding the level of measurement is important because it determines which statistical methods, calculations, and visualizations are appropriate for the data

Consider the variable customer satisfaction level (low, medium, high).
This is an ordinal level variable.

Finding the median makes sense because the data has an order.

Finding the mean does not make sense because the differences between categories are not equal.

If we wrongly treat this data as interval or ratio, the conclusions drawn would be invalid.

Measures of Central Tendency

Mean: Average value, shows overall center

Median: Middle value, useful with outliers

Mode: Most frequent value

Measures of Variability

Range: Difference between highest and lowest values

Variance: Spread of data around the mean

Standard Deviation: Average deviation from the mean

IQR: Spread of the middle 50% of data
