```python
import seaborn as sns
```

5 plots that we can plot in seaborn are
line graph-for seeing the trends
bar graph-for skew,for categorization.
scatter graph-to show the points in dots.For outlier
histogram-to show skew
box plot-for median 


```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
sns.lineplot(x=x,y=y)
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
sns.boxplot(x=x,y=y)
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
sns.scatterplot(x=x,y=y)
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
sns.barplot(x=x,y=y)
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
sns.histplot(x=x,y=y)
```

```python
fmri=sns.load_dataset('fmri')
sns.lineplot(x=fmri['timepoint'],y=fmri['signal'])
```

```python
#Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
#'pclass', y = 'age' and y = 'fare'.
titanic=sns.load_dataset('titanic')
sns.boxplot(x=titanic['pclass'],y=titanic['age'])
```

```python
titanic=sns.load_dataset('titanic')
sns.boxplot(x=titanic['pclass'],y=titanic['fare'])
```

```python
#Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue 
# parameter for the 'cut' column of the diamonds dataset.
diamonds=sns.load_dataset('diamonds')
sns.histplot(x=diamonds['price'],hue=diamonds['cut'])
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(data=iris, hue='species')
```

```python
flight = sns.load_dataset('flights')
sns.heatmap(flight.pivot(index='month', columns='year', values='passengers'))
```
