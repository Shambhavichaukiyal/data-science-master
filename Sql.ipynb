It is used to manipulate the data.

UPDATE
Purpose: Used to modify existing records in a table.

SET age = 23
WHERE id = 1;

Purpose: Used to remove records from a table.
DELETE FROM Student

Data query language -it is used to retrieve data from the database.

Primary key is a unique and not null in the table while foreign key is used to connect 2 table togther it is a primary key on other table.

CREATE TABLE Student (

CREATE TABLE Marks (
