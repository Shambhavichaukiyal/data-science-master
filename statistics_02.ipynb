the three measures of central tendency-
mean
median
mode

Mean: Average of the data

Median: Middle element of the data when arranged in order

Mode: Element with the highest frequency

```python
import numpy as np
from scipy import stats
data=[178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]
mean=np.mean(data)
median=np.median(data)
mode=stats.mode(data)
print("mean= ",mean)
print("median= ",median)
print("mode= ",mode)
```

```python
import numpy as np
data=[178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]
sta=np.std(data)
print(sta)
```

Covariance

Definition: Measures how two variables change together.

Interpretation:

Positive covariance: variables increase or decrease together

Negative covariance: one increases while the other decreases

Correlation

Definition: A standardized form of covariance that measures both strength and direction of a relationship.

Range: −1 to +1

+1 → perfect positive relationship

−1 → perfect negative relationship

0 → no relationship

Outliers are extreme values that are very different from the rest of the data. They can significantly distort measures of central tendency and dispersion.

Effect on Measures of Central Tendency

Mean:
Highly affected by outliers because it uses all values.
An extreme value pulls the mean toward itself.

Median:
Least affected by outliers since it depends on position, not magnitude.

Mode:
Usually not affected unless the outlier repeats frequently.

Example

Data: 10, 12, 13, 14, 15

Mean = 12.8

Median = 13

Now add an outlier 100:

New Data: 10, 12, 13, 14, 15, 100

Mean = 27.3 (increases sharply)

Median = 13.5 (slight change)

Range increases from 5 to 90

Variance & Standard Deviation increase significantly
