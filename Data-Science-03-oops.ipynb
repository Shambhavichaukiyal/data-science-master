multiple inheritance can be defined as a class acquiring property of other class in which A class acquire property of 2 or more class.

```python
class pwskills:
    def web(self):
        return "study web"
```

```python
class selfstudy:
    def dsa(self):
        return "study dsa"
```

```python
class student(pwskills ,selfstudy):
    def study(self):
        return "study"
```

```python
obj=student()
```

```python
obj.study()
```

```python
obj.web()
```

class.

Getters and Setters are methods used to access and modify the variables of a class.

```python
class test:
    def set_marks(self,marks):
        self._marks=marks
    def get_marks(self):
        return self._marks
    
```

```python
obj=test()
```

```python
obj.set_marks(50)
```

```python
obj.get_marks()
```
