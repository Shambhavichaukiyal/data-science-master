Multithreading can be defined as executing more than one thread run concurrently.
threading is the module used to handle threads in python

thread module is used to work with thread synchronise them, execute them and what not.

activeCount()-It is used to count number of thread that are alive.

currentThread-It makes the object of current thread.

enumerate-gives list of all thread thar alive.



```python
import threading
print(threading.activeCount())
```

```python
import threading 
print(threading.currentThread().name)
```

```python
import threading

all_threads = threading.enumerate()
print("All threads currently active:")
for thread in all_threads:
    print(thread.name)
```

run()-It is a method which reside where actual code the thread works on reside.
start()-It means to execute the thread.
join()-It is used to wait till the thread get executed.
isalive()-if the thread is currently runnning or not.

```python
import threading

def print_squares():
    squares = [x ** 2 for x in range(1, 11)]
    print("List of squares:", squares)

def print_cubes():
    cubes = [x ** 3 for x in range(1, 11)]
    print("List of cubes:", cubes)


thread1 = threading.Thread(target=print_squares)


thread2 = threading.Thread(target=print_cubes)


thread1.start()
thread2.start()


thread1.join()
thread2.join()

print("Both threads have finished execution.")
```

Advantage of MultiThreading-

Disadvantage of Multithreading

Complexity

Difficulty in debugging

Deadlock in a multithreaded program is where one process is waiting for a resource but other process is also waiting fro a resource which is with process 1.
Race conditions occur when the outcome of a program depends on the timing or interleaving of operations by multiple threads.
