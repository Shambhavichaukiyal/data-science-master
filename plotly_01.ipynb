```python
#Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.
import plotly.graph_objects as go
import plotly.io as pio
import seaborn as sns
```

```python
pio.renderers.default='browser'
titanic=sns.load_dataset('titanic')
fig=go.Figure()
fig.add_trace(go.Scatter(x=titanic['age'],y=titanic['fare'],mode='markers'))
```

```python
#Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

tips=sns.load_dataset('tips')
fig=go.Figure()
fig.add_trace(go.Box(x=tips['total_bill']))
```

```python
#Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"column with the color parameter.
```

```python
import plotly.express as px
import seaborn as sns

tips = sns.load_dataset('tips')

fig = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    pattern_shape="smoker",
    color="day"
)

fig.show()
```

```python
import plotly.express as px

# Load iris dataset from Plotly
iris = px.data.iris()

# Scatter matrix plot
fig = px.scatter_matrix(
    iris,
    color="species",
    title="Scatter Matrix Plot of Iris Dataset"
)

fig.show()
```

```python
#A Distplot (Distribution Plot) is used to visualize the distribution of a continuous numerical variable.
import plotly.express as px

# Sample data (using iris dataset)
iris = px.data.iris()

# Distplot
fig = px.histogram(
    iris,
    x="sepal_length",
    marginal="rug",
    title="Distplot of Sepal Length"
)

fig.show()
```
