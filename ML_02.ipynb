Overfitting can be defined as when accuracy in trained dataset is high but accuracy in unknown dataset is low.
Consequences-
low bias
high variance
high accuracy in training dataset
low accuracy in test/validation dataset
Mitigated-
Increase the dataset

Consequences-
high bias
low variance
low accuracy in training dataset
low accuracy in test/validation dataset
Mitigated-
Increase the dataset 
less regularization
include more features
